```python
from keras.datasets import mnist
import numpy as np
import tensorflow as tf
import pandas as pd
import os
import csv
from sklearn.model_selection import train_test_split

#path_photos = '../../archive/img_align_celeba/img_align_celeba'
#path_labels = '../../archive/list_attr_celeba.csv'

path_photos = 'Ressources_Test\images'
path_labels = 'Ressources_Test\img_attributes.csv'

labels_dic = pd.read_csv(path_labels)
#print(labels_dic)
y = []

for row in labels_dic.iterrows():
    #print("row",row[1][0])
    sub_y = []
    for i in range(1,len(row[1])) :
        sub_y.append(row[1][i])
    #print(sub_y)
    y.append(sub_y)

#print(y)
"""
(x_train, y_train), (x_test, y_test) = mnist.load_data(path='../../archive/img_align_celeba/img_align_celeba')


x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.
"""

x = []


for filename in os.listdir(path_photos):
    
    f = os.path.join(path_photos, filename)
    # checking if it is a file
    if os.path.isfile(f):
        #print(f)
        image = tf.keras.preprocessing.image.load_img(f)
        input_arr = tf.keras.preprocessing.image.img_to_array(image)
        #print("image to array : ",input_arr)
        
        x.append(input_arr)
        

#input_arr = np.array([input_arr])  # Convert single image to a batch.
#print("convert to batch", input_arr)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33)

#x_train = x_train / 255.
#x_test = x_test / 255.

#print('x data shape: ', x)
#print('y data shape: ', y)

x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
y_test = np.array(y_test)

print(x_train)
print('Training data shape: ', x_train.shape)
print('Testing data shape : ', x_test.shape)
print('Training labels shape: ', y_train.shape)
print('Testing labels shape : ', y_test.shape)
print(y_train)
```

```python
from __future__ import print_function
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
from keras.layers import Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint
from keras.models import model_from_json
from keras import backend as K
```

```python
# The known number of output classes.
num_classes = 10

# Input image dimensions
img_rows, img_cols = 218, 178


# Channels go last for TensorFlow backend
x_train_reshaped = x_train.reshape(-1, img_rows, img_cols, 3)
x_test_reshaped = x_test.reshape(-1, img_rows, img_cols, 3)
input_shape = (img_rows, img_cols, 3)

print(len(x_train_reshaped))

# Convert class vectors to binary class matrices. This uses 1 hot encoding.
y_train_binary = keras.utils.to_categorical(y_train, num_classes)
y_test_binary = keras.utils.to_categorical(y_test, num_classes)

print(y_train[0])
print(y_train_binary[0])
```

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), padding='same', input_shape=(28, 28, 3)))
model.add(Activation('relu'))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(num_classes))
model.add(Activation('softmax'))

model.compile(loss='categorical_crossentropy', optimizer='Adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10)
```
